### TensorFlow Dense Layer

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers, losses, optimizers, metrics, Sequential

import numpy as np
```

```python
x = np.array([-1, 0, 1, 2, 3, 4], dtype = float)
y = np.array([-3, -1, 1, 3, 5, 7], dtype = float)
```

```python
model = Sequential([
    layers.Dense(1, input_shape=(1,))
])

loss_function = losses.MeanAbsoluteError()
optimizer = optimizers.Adam(learning_rate=0.01)
metric = metrics.MeanAbsoluteError()

model.compile(loss=loss_function, optimizer=optimizer, metrics=[metric])
model.summary()
```

```python
model.fit(x, y, epochs = 500)
```

```python
print(model.predict(x).reshape(1, -1))
print(y)
```

### From scratch with GradientTape

```python
w = tf.Variable(5.0, trainable = True)
b = tf.Variable(10.0, trainable = True)
```

```python
def loss_function(y, y_pred):
    return tf.reduce_sum(tf.abs(y - y_pred))

learning_rate = 0.01
epochs = 2000
```

```python
"""
When persistent=True is set, the GradientTape object allows multiple calls to gradient()
before needing to be explicitly deleted.
"""

def train(x, y):
    with tf.GradientTape(persistent = True) as tape:
        tape.watch([w, b])
        y_pred = w * x + b
        loss = loss_function(y, y_pred)

    dw, db = tape.gradient(loss, [w, b])

    w.assign_sub(learning_rate * dw)
    b.assign_sub(learning_rate * db)
    del tape
    return loss
```

```python
for epoch in range(epochs + 1):
    loss = train(x, y)
    if epoch % 100 == 0:
        print("Epoch", epoch, " ==> Loss", loss.numpy(), " ==> Paramaters (w, b)", w.numpy(), b.numpy())
```

```python
print("Predictions", w.numpy() * x + b.numpy())
print("Actual values",  y)
```

### Analyze Gradient Tape

```python
k = tf.Variable(2.0)

with tf.GradientTape() as tape:
    h = k ** 2 + 5 * k - 10

print(tape.gradient(h, k))
```

```python
k = tf.Variable(2.0)
g = tf.Variable(3.0)

with tf.GradientTape() as tape:
    h = k ** 2 + 5 * k * g - 3 * g

for grad in tape.gradient(h, [k, g]):
    print(grad.numpy())
```
