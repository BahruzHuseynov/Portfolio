```python
import tensorflow as tf
```

### tf.Variable (Can be modified)

```python
print(tf.Variable([2, 3], dtype = tf.int32))
```

```python
# Note that 'dtype' must be mentioned. It is not the second parameter in the method
# The problem: After executing the cell, you will see "dtype" still equal to "int32"
tf.Variable([2, 3], tf.float32)
```

```python
# Additionally, the shape of the tensor is not changable with tf.Variable()
tf.Variable([1,2,3,4], shape=(2,2))
```

### Other Data Types

```python
print(tf.Variable([2, 3], dtype = tf.float32))
print(tf.Variable([1, 2, 3, 4, 5], dtype = tf.int32))
print(tf.Variable(3 + 2j, dtype = tf.complex64))
print(tf.Variable("TF", dtype = tf.string))
```

### tf.constant (Cannot be modified)

```python
# No problem about "shape"
# Same problem about "dtype"
print(tf.constant([5, 8], dtype = tf.int32))
print(tf.constant([1, 2, 3, 4], shape = (2, 2)))
print(tf.constant(4, shape = (3, 2), dtype = tf.float32))
```

### Math Operations

```python
a = tf.Variable([[1, 2], [3, 4]])
b = tf.Variable([[1, 1], [1, 1]])
c = tf.Variable([[5], [3]])
```

```python
print(a + b)
print(tf.add(a, b))
```

```python
print(a-b)
print(tf.subtract(a, b))
```

```python
print(a @ b)
print(tf.matmul(a, b))
```

```python
print(a / 2)
print(tf.divide(a, 2))
```

```python
print(tf.square(5))
print(tf.sqrt(16.0))
print(tf.reduce_sum(a))
print(tf.reduce_max(a))
print(tf.reduce_min(a))
print(tf.reduce_mean(a))
```

### Type Conversion

```python
x = tf.Variable([2, 3], dtype = tf.float32)
x = tf.cast(x, tf.int32)
print(x)
```

### Shape Operation and Tensor Manipulation

```python
print(tf.shape(x))
print(tf.size(x))
print(tf.reshape(x, (1, 2)))
```

```python
arr = tf.Variable([[1, 2], [3, 4]])
print(tf.transpose(arr))
```

```python
r = tf.Variable([[[1, 2], [3, 4]]])
print(tf.shape(r))
print(tf.squeeze(r))
print(tf.expand_dims(r, 1))
```

```python
x = tf.Variable([[1, 3], [2, 4]])
print(tf.boolean_mask(x, x > 2))
```
