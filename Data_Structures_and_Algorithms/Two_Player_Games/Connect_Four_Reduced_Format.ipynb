# Connect Four - Reduced Format (Minimax with manipulating the depth)

<p style = "font-size: 18px"> Instead of using 6x6 Connect Four board, we have reduced the shape of the board to 4x4, and in order to win the game the sequence of 3 same figures must be obtained
</p>

```python
import numpy as np
```

```python
def find_winner(board):
    # Checking columns
    for i in range(4):
        if "xxx" in "".join(list(board[:, i])):
            return -1
        if "ooo" in "".join(list(board[:, i])):
            return 1
    
    # Checking rows
    for row in board:
        if "xxx" in "".join(row):
            return -1
        if "ooo" in "".join(row):
            return 1
    
    # Checking diagonals
    for i in range(2):
        l1 = "".join(list(board[np.arange(0, 4-i), np.arange(i, 4)]))
        l2 = "".join(list(board[np.arange(i, 4), np.arange(0, 4-i)]))
        l3 = "".join(list(board[np.arange(0, 4-i), np.arange(0, 4-i)[::-1]]))
        l4 = "".join(list(board[np.arange(i, 4), np.arange(i, 4)[::-1]]))
        
        # print(l1, l2, l3, l4, end = "\n")
        if "xxx" in l1 or "xxx" in l2 or "xxx" in l3 or "xxx" in l4:
            return -1
        if "ooo" in l1 or "ooo" in l2 or "ooo" in l3 or "ooo" in l4:
            return 1
    
    # Filled or not
    if " " in board:
        return "game continues"
    
    # Draw
    return 0
```

```python
def minimax_alpha_beta(board, turn, depth = 3, alpha = -float("inf"), beta = float("inf")):
    # print(board)
    winner = find_winner(board)
    if winner == 1 or winner == -1:
        return winner
    
    if depth == 0:
        return 0
    
    if turn:
        best_score = -float("inf")
        best_col = None
        for col in range(4):
            row = "".join(list(board[:, col])).rfind(" ")
            if row != -1:
                board[row][col] = "o"
                score = minimax_alpha_beta(board, False, depth - 1, alpha, beta)
                board[row][col] = " "

                if type(score) is tuple:
                    score = score[0]

                if score > best_score:
                    best_score = score
                    best_col = col
                
            
        return best_score, best_col
    else:
        best_score = float("inf")
        worst_col = None
        for col in range(4):
            row = "".join(list(board[:, col])).rfind(" ")
            if row != -1:
                board[row][col] = "x"
                score = minimax_alpha_beta(board, True, depth - 1, alpha, beta)
                board[row][col] = " "
                    
                if type(score) is tuple:
                    score = score[0]
                    
                if score < best_score:
                    best_score = score
                    worst_col = col
                
                    
        return best_score, worst_col
```

```python
# Testing
board = np.array([[" "," "," "," "],
                 [" "," "," ","o"],
                 [" "," ","x","x"],
                 ["o","o","x","x"]])

# depth = np.count_nonzero(board == " ")
best_score, best_col = minimax_alpha_beta(board, True)
print(best_score, best_col)
```

# GAME - REDUCED FORMAT OF CONNECT FOUR

```python
def play_your_turn(board, whose_turn, col):
    x = board[:, col][::-1]
    x["".join(x).find(" ")] = whose_turn
    board[:, col] = x[::-1]
```

```python
def print_cf_board(board):
    print("\n-----------------")
    for row in board:
        print("|", end = "")
        for col in row:
            print(f" {col} ", end = "|")
        print("\n-----------------")
```

```python
represent = np.array([["0","1","2","3"],
                     [" "," "," "," "],
                     [" "," "," "," "],
                     [" "," "," "," "],
                     [" "," "," "," ",]])
represent
```

```python
board = np.array([[" "," "," "," "],
                [" "," "," "," "],
                [" "," "," "," "],
                [" "," "," "," ",]])

print_cf_board(represent)
while np.count_nonzero(board == " ") != 0:
    me = int(input("Your turn. Play: "))
    if str(me) not in "0123":
        print("Choose correct number to play !!!")
        continue
    play_your_turn(board, "x", me)
    print_cf_board(board)
    if find_winner(board) == -1:
        break
    
    # AI will decide the movement
    # depth = np.count_nonzero(board == " ")
    ai = minimax_alpha_beta(board, True)
    play_your_turn(board, "o", ai[1])
    print_cf_board(board)
    
    if find_winner(board) == 1:
        break
        
winner = find_winner(board)
if winner == -1:
    print("You won")
elif winner == 1:
    print("Computer won")
else:
    print("Draw!")
```
