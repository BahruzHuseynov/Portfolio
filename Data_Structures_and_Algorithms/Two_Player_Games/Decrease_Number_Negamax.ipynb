```python
def find_winner(val, player):
    if val <= 0:
        return not player
    return 0
```

```python
def negamax(val, comp, depth, player):
    winner = find_winner(val[0], player)
#     print("---------------")
#     print(val)
#     print("Winner: ", winner)
#     print("Player: ", player)
    if winner != 0:
        return winner, 0
    
    if depth == 0:
        return 0, 0
    
    bestMove = 0
    bestEvaluation = -float('inf')
    res = val[0]
    
    
    for i in range(1,3):
        if i == 1:
            val[0] = val[0]//2
            evaluateMove = -int(negamax(val, comp, depth - 1, not player)[0])
#             print("++++++++++++++++")
#             print(evaluateMove)
            val[0] = res
            if evaluateMove > bestEvaluation:
                bestMove = 1
                bestEvaluation = evaluateMove

        elif i == 2:
            val[0] = val[0] - comp
            evaluateMove = -int(negamax(val, comp, depth - 1, not player)[0])
            val[0] = res
            if evaluateMove > bestEvaluation:
                bestMove = 2
                bestEvaluation = evaluateMove

    return evaluateMove, bestMove
```

# GAME - Decrease the number

### Decrease - 1
### Divide - 2

```python
val = random.randrange(50,100)
dec = random.randrange(1,10)

print("General number:", val)
print("Choice to decrease general number: ", dec)

while val > 0:
    choice = int(input(f"Decrease the number by {dec} or divide by 2?: "))
    if choice not in [1,2]:
        print("Make a right choice!!!")
        continue
    
    if choice == 1:
        val = val - dec
    elif choice == 2:
        val = val // 2
    
    print(f"You decreased number to {val}")
    if val <= 0:
        print("Computer won")
        break
    
    print("------------------------------------------------------------------")
    # Computer will play
    depth = max(val // choice, val // math.floor(math.log2(val)))
    comp_choice = negamax([val], dec, depth, True)
    if comp_choice[1] == 2:
        print(f"Computer's choice is 1")
        val = val // 2
    elif comp_choice[1] == 1:
        print(f"Computer's choice is 2")
        val = val - dec
    
    print(f"Computer decreased number to {val}")
    
    if val <= 0:
        print("You won")
        break
        
    print("------------------------------------------------------------------")
```
