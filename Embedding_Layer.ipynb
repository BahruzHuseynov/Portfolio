# Embedding Layer

The Embedding layer in neural networks for text data converts integer-encoded words into dense vectors during training. These vectors are learned and updated through backpropagation, refining their values based on the model's learning from the data. This transformation enables the model to capture semantic relationships and contextual meanings of words, enhancing its ability to perform well on tasks like natural language processing and sentiment analysis. <br>

As the model learns from the training data, the weights (which include the values in the embedding vectors) are adjusted to minimize the loss function, improving the model's ability to make accurate predictions. Therefore, through backpropagation, the values in the embedding vectors are updated iteratively to better represent the relationships between words learned from the data.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding

# Parameters
vocab_size = 10000    # Number of unique words in the vocabulary
embedding_size = 128  # Dimension of the embedding vector
maxlen = 500          # Maximum length of input sequences

# Build the model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_size, input_length=maxlen))

# Print model summary
model.summary()
```

```python
import torch
import torch.nn as nn

# Parameters
vocab_size = 10000    # Number of unique words in the vocabulary
embedding_size = 128  # Dimension of the embedding vector

# Define the embedding layer
embedding_layer = nn.Embedding(num_embeddings=vocab_size, embedding_dim=embedding_size)

# Print the embedding layer
print(embedding_layer)
```
