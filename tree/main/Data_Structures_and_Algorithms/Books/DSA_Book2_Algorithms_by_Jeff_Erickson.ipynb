```python
def Lattice_Multiplication(X, Y):
  X, Y = X[::-1], Y[::-1]
  lst = []
  for i in range(len(X)):
    s = " " * i
    hold = 0
    for j in range(len(Y)):
        hold += int(X[i]) * int(Y[j])
        print(hold)
        s += str(hold % 10)
        hold //= 10
    if hold != 0:
        s += str(hold)
    lst.append(s)

  return lst
```

```python
Y = "23"
X = "67"
print(Lattice_Multiplication(X, Y))
```

```python
def Peasant_Multiplication(x, y):
    prod = 0
    while x > 0:
        if x % 2 == 1:
            prod += y
        x = x // 2
        y = y + y
    return prod
```

```python
x = 60
y = 27
print(x * y)
print(Peasant_Multiplication(x, y))
```
