### Introduction to Tensors

```python
# Scalar
scalar = torch.tensor(7)
print(scalar)
print(scalar.ndim)
print(scalar.shape)
print(scalar.item())
```

```python
# Vector
vector = torch.tensor([2,-5, 10, 7])
print(vector)
print(vector.ndim)
print(vector.shape)
print(vector.numpy())
```

```python
# Matrix
matrix = torch.tensor([[2,-5], [10, 7]])
print(matrix)
print(matrix.ndim)
print(matrix.shape)
print(matrix.numpy())
```

```python
# Tensor
t = torch.tensor([[[1,2,3],
                   [3,6,9],
                   [2,4,5]],
                  [[3,1,0],
                   [5,2,4],
                   [2,9,8]]])
print(t)
print(t.ndim)
print(t.shape)
print(t.numpy())
```

```python
# Random tensors
random_tensor = torch.rand(3, 4)
print(random_tensor)
print(random_tensor.ndim)
print(random_tensor.shape)

print("--------------------------")
# Random tensors
random_tensor = torch.rand(2, 5, 5)
print(random_tensor)
print(random_tensor.ndim)
print(random_tensor.shape)
```

```python
# Create a random tensor of size (224, 224, 3) like RGB
random_image_size_tensor = torch.rand(size=(3, 224, 224))
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

```python
# Zeros and Ones
print(torch.zeros(2,2))
print(torch.ones(size = (3,4)))
```

```python
# Data Type
t = torch.rand(size = (2,3))
print(t.dtype)
print(t)
print("-----------------------------------")
t = t.to(torch.int32)
print(t.dtype)
print(t)
```

```python
# Range
rng = torch.arange(1, 5)
print(rng)
print("------------------------------")
rng = torch.arange(1, 10, 3)
print(rng)
```

```python
# Tensors like
three_zeros = torch.zeros_like(input = rng)
three_zeros
```

### Manipulating tensors

```python
# Tensor and Scalar
t = torch.tensor([5,10,3])
print(t+10) # torch.add(t, 10)
print("-------------------------")
print(t*10) # torch.mul(t, 10)
print("-------------------------")
print(t-10) # torch.subtract(t, 10)
print("-------------------------")
print(t//10)
print("-------------------------")
print(t/10) # torch.divide(t, 10)
```

```python
# Matrix multiplication
t = torch.tensor([1,2,3])

print(t*t)
print("-------------------------------")
print(torch.matmul(t, t))
print("-------------------------------")
print(t @ t)
```

```python
a = torch.tensor([[1,2],
                  [3,4]])

b = torch.tensor([[5,10,-8,0],
                  [9,7,5,-2]])
print(a @ b)
```

```python
print(a.T) # Transpose
print(b.T)
```

```python
# Tensor aggregation (min, max, sum, mean)
x = torch.arange(0, 100, 10)
print(x)
print("----------------------")
print(torch.min(x), x.min())
print(torch.max(x), x.max())
print(torch.sum(x), x.sum())
print("----------------------")
x = x.type(torch.float32)
print(torch.mean(x), x.mean())
```

```python
print(x.argmin(), torch.argmin(x))
print(x.argmax(), torch.argmax(x))
```

```python
x = torch.arange(1,10)
x, x.shape
```

```python
# Reshaping
x_reshaped = x.reshape(3,3)
x_reshaped, x_reshaped.shape
```

```python
# Viewing
z = x.view(3, 3) # z will share the same memory with x
z, z.shape

# so, if a value (not a shape) in 'z' changes, it will change in the same manner in 'x'
z[0, 1] = 10
print(z, x, sep = "\n")
```

```python
# Stacking
x_stacked = torch.stack([x,x,x]) # default dim = 0
x_stacked
```

```python
x_stacked = torch.stack([x,x,x], dim = 1)
x_stacked
```

```python
# Squeezing and Unsqueezing
# Squeezing removes all single dimensions
x = torch.rand(size = (3,1,2,1))
print(x)
print("------------------------------")
print(torch.squeeze(x))
print("------------------------------")
print(x.squeeze())
```

```python
# Unsqueezing - adds a single dimension to a target tensor at a specific dimension
x = torch.rand(size = (3,2))
print(x)
print(x.unsqueeze(dim = 0))
```

```python
# Permute
x = torch.rand(size = (2, 4, 3))
y = x.permute(2, 0, 1)
print(x.shape)
print(y.shape)
print("-----------------------------------")
print(x)
print(y)
```
