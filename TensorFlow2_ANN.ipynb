### Test 1

```python
import numpy as np

import tensorflow as tf
from tensorflow import keras
from keras import Sequential, layers
```

```python
X_train = np.array([-3, -1,  0, 1, 2, 3, 4])
y_train = np.array([-5, -3, -1, 1, 3, 5, 7])
X_test = np.array([-2, 5, -4 ])
y_test = np.array([-4, 9, -7])
```

```python
# Model Building
model = Sequential([
    layers.Dense(1, input_shape=[1])
])
model.compile(optimizer='sgd', loss = 'mean_squared_error', metrics = ['mae'])
model.fit(X_train, y_train, epochs = 250, validation_data = (X_test, y_test))
```

```python
# Predictions
preds = model.predict(X_test)
print(preds[:, 0])
print(y_test)
```

### Test 2 - Approximately, y = 2x + 1

```python
X_train = np.array([4, 5, 8, 9, 12, 20, 32, 84])
y_train = np.array([9, 10, 17, 19, 24, 40, 64, 170])
X_test = np.array([6, 10, 14, 24])
y_test = np.array([13, 21, 29, 49])
```

```python
# Model Building
# Loss increases to INFINITY
model = Sequential([
    layers.Dense(1, input_shape=[1])
])
model.compile(optimizer='sgd', loss = 'mean_squared_error')
model.fit(X_train, y_train, epochs = 15)
```

```python
# Let's change the optimizer from SGD to ADAM
model = Sequential([
    layers.Dense(1, input_shape=[1])
])
model.compile(optimizer='adam', loss = 'mean_squared_error', metrics = ['mae'])
model.fit(X_train, y_train, epochs = 500, validation_data = (X_test, y_test))
```

```python
# The result above is not satisfactory, too. We can use optimizers, losses from tensorflow
# Then play with the parameters
from keras.optimizers import Adam
from keras.losses import MeanSquaredError
from keras.metrics import MeanAbsoluteError

model = Sequential([
    layers.Dense(1, input_shape=[1])
])
model.compile(optimizer=Adam(learning_rate=0.1), loss = MeanSquaredError(), metrics = [MeanAbsoluteError()])
model.fit(X_train, y_train, epochs = 500, validation_data = (X_test, y_test))
```

```python
model.get_weights() # k = 2.00252268, b = 0.94068134
```

```python
preds = model.predict(X_test)
preds[:, 0], y_test
```
