```python
class Stack:
  def __init__(self):
    self.items = []

  def is_empty(self):
    return self.items == []

  def push(self, item):
    self.items.append(item)

  def pop(self):
    return self.items.pop()

  def peek(self):
    return self.items[len(self.items)-1]

  def size(self):
    return len(self.items)

  def display(self):
    for item in self.items[::-1]:
      print(item)
```

```python
s = Stack()
print(s.is_empty())
s.push(4)
s.push('dog')
print(s.peek())
s.push(True)
print(s.size())
print(s.is_empty())
s.push(8.4)
print(s.pop())
print(s.pop())
print(s.size())
```

```python
s.display()
```

```python
def dec_to_bin(dec):
  ans = Stack()

  while dec > 0:
    ans.push(dec % 2)
    dec //= 2

  s = ""
  while not ans.is_empty():
    s += str(ans.pop())

  return s
```

```python
print(dec_to_bin(40))
```

```python
def base_converter(dec, base):
  digits =  "0123456789ABCDEF"
  ans = Stack()

  while dec > 0:
    rem = dec % base
    ans.push(digits[rem])
    dec //= base

  s = ""
  while not ans.is_empty():
    s += str(ans.pop())

  return s
```

```python
base_converter(28, 16)
```

## Queues

<p style = "font-size: 18px"> A queue is a collection of items where new items are added at one end (rear) and existing items are removed from the other end (front). This follows the FIFO (First-In, First-Out) principle, meaning the item added earliest is removed first. Queues are commonly seen in everyday scenarios like waiting in lines at movie theaters or grocery stores, and are also used in computer science for tasks like printer queues and process scheduling. They ensure orderly processing by adhering to a strict "first-come, first-served" approach, allowing tasks to be completed in the order they were initiated. </p>

```python
class Queue:
  def __init__(self):
    self.items = []

  def is_empty(self):
    return self.items == []

  def enqueue(self, item):
    self.items.append(item)

  def dequeue(self):
    return self.items.pop(0)

  def size(self):
    return len(self.items)

  def display(self):
    print(self.items)
```

```python
q = Queue()
q.enqueue('hello')
q.enqueue('dog')
q.display()
q.enqueue(3)
q.dequeue()
q.display()
```
